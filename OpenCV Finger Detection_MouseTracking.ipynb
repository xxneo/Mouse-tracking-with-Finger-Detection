Import Packages

```python
import cv2 as cv
import mediapipe as mp
import pyautogui
```

```python
#Open WebCam
cap = cv.VideoCapture(0, cv.CAP_DSHOW)

#setting up mediapipe for hamd detector
hand_detector = mp.solutions.hands.Hands()

#creating an instance of MediaPipe drawing
mp_draw = mp.solutions.drawing_utils

while cap.isOpened():
    ret, frame = cap.read()
    
    #Converting to colour screen
    rgb_frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    output = hand_detector.process(rgb_frame)
    hands = output.multi_hand_landmarks
    print(hands)
    
    if hands:                     #if hands!=None
        for hand in hands:
            mp_draw.draw_landmarks(frame, hand)
    
    #the next line draws all 21 landmarks for the current hand
    

    cv.imshow("Frame",frame)
    #if user presses 'q' stop video capture

    #0xFF is ASCII for 'Q'
    if cv.waitKey(5) & 0xFF == ord('q'):
        break
        
cap.release()
cv.destroyAllWindows()
    
    
```
